### Cleaning data

```python
#axix = 0 rows
#axix = 1 col
nan_df = data[data.isna().any(axis=1)]
nan_df.head(100)
```

```python
#how='any'
data.dropna(how='all', inplace=True)
data.head()
```

```python
data.info()
```

#### what was the best month for sales?

```python
#step1:
data = data[data['Order Date'].str[0:2] != 'Or']
```

```python
data['Month'] = data['Order Date'].str[0:2]
```

```python
data.tail()
```

```python
data['Month'] = data['Month'].astype('int32')
```

```python
data.head()
```

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])

data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head(100)
```

```python
results = data.groupby('Month').sum()
```

```python
results
```

```python
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.show()
```

##### Which city has the highest no. of sales?

```python


data['City'] = data['Purchase Address'].apply(lambda x : x.split(',')[1] + ' ' + x.split(',')[2].split(' ')[1])
data.head()
```

```python
results = data.groupby('City').sum()
results
```

```python
cities = [city for city, df in data.groupby('City')] #list comprihension for putting cities in order

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Cities')
plt.show()
```

##### what time should we display the advertisments to maximize likelihood of customer's buying product?

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.head()
```

```python
data['Hour'] = data['Order Date'].dt.hour
data.head()
```

```python
data['Minute'] = data['Order Date'].dt.minute
data.head()
```

```python
hours = [hour for hour, df in data.groupby('Hour')]

plt.plot(hours, data.groupby('Hour').count())
plt.xticks(hours)
plt.grid()
plt.show()
```

```python
# So good time to put our ads is before 11am and after 5pm
```
